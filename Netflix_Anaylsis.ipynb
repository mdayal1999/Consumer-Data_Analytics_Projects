# Distribution of Ratings

```python
rating_counts = df['rating'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
rating_counts.plot(kind='bar', color='mediumpurple')
plt.title('Distribution of Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- The plot reveals that the most common content ratings on Netflix are TV-MA, TV-14, and TV-PG, indicating a strong focus on mature and teen audiences. There’s comparatively less content for younger children (like TV-Y or G). This suggests Netflix prioritizes content for adult and young adult viewers, aligning with its global audience's preferences for drama, thrillers, and complex narratives.

# Spliting duration based on Content Type

```python
df_movies = df[df['type'] == 'Movie'].copy()
df_movies['duration_mins'] = df_movies['duration'].str.extract('(\d+)').astype(float)

df_tv = df[df['type'] == 'TV Show'].copy()
df_tv['seasons'] = df_tv['duration'].str.extract('(\d+)').astype(float)
```

- To analyze movies and TV shows separately, we created two DataFrames: one for movies and one for TV shows. Since the duration column contains text like "90 min" or "2 Seasons", we extracted the numeric values and converted them to float. This allows us to perform meaningful analysis, such as calculating average movie lengths or examining the distribution of TV show seasons, enabling clearer insights into content structure on Netflix.



# Distribution of Movie Durations

```python
plt.figure(figsize=(10, 6))
plt.hist(df_movies['duration_mins'].dropna(), bins=30, color='salmon', edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

- Most movies on Netflix tend to have durations between 80 to 100 minutes, indicating that standard-length films are the most common. There are fewer movies that are very short (under 60 minutes) or very long (over 120 minutes), suggesting a typical runtime preference among Netflix’s movie catalog.

This helps understand content length trends and can inform recommendations or platform strategy (e.g., what duration ranges are most prevalent).

# TV Shows: Number of Seasons

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df_tv, x='seasons', color='teal')
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

As the number of seasons increases, the count of shows decreases significantly. This trend suggests that Netflix either frequently acquires or produces many short-running series, or that longer-running shows are less frequent in its catalog.

This can reflect content strategy (focus on fresh, short series) or viewer preferences for bingeable, single-season shows.

# Rating Trend Over Time

```python
rating_trend = df.groupby(['year_added', 'rating']).size().unstack(fill_value=0)

top_ratings = rating_trend.sum().sort_values(ascending=False).head(4).index
rating_trend[top_ratings].plot(figsize=(12, 6))
plt.title('Content Ratings Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```

- This line plot tracks the distribution of the top 4 content ratings on Netflix over the years, showing how the number of titles with each rating has evolved. It reveals trends such as the increasing presence of higher ratings (e.g., TV-MA, PG-13), suggesting a shift towards more mature content as Netflix expanded its library. Conversely, lower ratings may have stabilized or declined, reflecting changes in Netflix’s programming strategy to target specific audience demographics. This insight highlights how the platform's content offerings have evolved to cater to varying viewer preferences over time.
